```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../data/query_output/query_6.csv', index_col=False)

df['STADIUM'] = df['STADIUM'].str.replace('fifa world cup', 'FWC', case=False)

df.set_index('STADIUM', inplace=True)

df
```

```python
# Grouped bar plot
df[['NUMBER_OF_MATCH', 'NUMBER_OF_GOAL', 'GOAL_PER_MATCH']].plot(kind='bar', 
                                                               figsize=(14, 4),
                                                               color=['#2C4AC7', '#A72CBA', '#A6BA54'])

# find max_y values
max_nm = max(df.NUMBER_OF_MATCH)
max_ng = max(df.NUMBER_OF_GOAL)
max_gpm = max(df.GOAL_PER_MATCH)

max_value = max([max_nm, max_ng, max_gpm])

# Set plot title and labels
plt.title('Fifa World Cup 2006 - Stadium and their Match Statistics')
plt.xlabel('Stadium Name')
plt.ylabel('Count')

# Set y-axis ticks
plt.yticks(range(0, max_value + 1, 2)) 

# Rotate x-axis labels
plt.xticks(rotation=80)

plt.savefig('../Visuals/plots/query_6.png', bbox_inches='tight')

# Display the plot
plt.show()
```
