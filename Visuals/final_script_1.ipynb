```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../data/query_output/query_5_1.csv')
```

```python
df.head()
```

```python
# Plotting a box plot for both 'Value1' and 'Value2'
df.boxplot(column=['TOTAL_CARD', 'NUMBER_OF_GOAL'], grid=False, notch=True, vert=False, patch_artist=True)

# Set plot title and labels
plt.title('Box Plot of Total Cards & Total Goal')
plt.xlabel('Values')
plt.yticks([1, 2], ['TOTAL_CARD', 'NUMBER_OF_GOAL'])

max_value = max(max(df['TOTAL_CARD']), max(df['NUMBER_OF_GOAL']))
# Set y-axis ticks
plt.xticks(range(0, max_value + 1, 2))

plt.savefig('../Visuals/plots/query_5_1.png', bbox_inches='tight')

# Show the plot
plt.show()
```

```python
plt.scatter(df['TOTAL_CARD'], df['NUMBER_OF_GOAL'])

# Add labels and title
plt.xlabel('TOTAL_CARD')
plt.ylabel('NUMBER_OF_GOAL')
plt.title('Visual of Total Goals and Total Number of Cards in Each Match')

max_value = max(df['TOTAL_CARD'])
# Set y-axis ticks
plt.xticks(range(0, max_value + 1, 2)) 

plt.savefig('../Visuals/plots/query_5_2.png', bbox_inches='tight')

# Show the plot
plt.show()
```

```python
print(f'Correlation between goal and card: {df.NUMBER_OF_GOAL.corr(df.TOTAL_CARD)}')
```
